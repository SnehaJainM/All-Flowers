```python
dataframe=dataframe.rename(columns={0:'First',1:'Second',2:'Third',3:'Fourth',4:'Fifth'})
dataframe
# First	Second	Third	Fourth	Fifth
# 0	12.0	89	200.0	728.0	False
# 1	788.0	201	289.0	1001.0
# 2	NaN	29100	NaN	279.0	211
# 3	100.0	290	299.0	NaN	23
```

```python
# dataframe['First'].strip(' ')

# AttributeError: 'Series' object has no attribute 'strip'
dataframe['First'].str.strip(' ') #Trailing and ending
```

```python
# dataframe['Second'].replace([1001,101],[1000,100])
dataframe['Second']=dataframe['Second'].replace(1000,101)
dataframe
```
